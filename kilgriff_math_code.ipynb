```python
data = pd.read_csv('corpus_data_all.csv').drop('Unnamed: 0', 1)
data.head()
```

```python
from collections import counter
```

```python
count = Counter(data['word'])
```

```python
import operator
sorted_count = sorted(count.items(), key=operator.itemgetter(1), reverse=True)
sorted_count
```

```python
freqs = [a[1] for a in sorted_count]
freqs
```

```python
count_freqs = sorted(Counter(freqs).items(), key=operator.itemgetter(0))#, reverse=True)
count_freqs
```

```python
df = pd.DataFrame(count_freqs, columns=['Freq','N'])
df
```

```python
def smooth(n, values):
    new_list = ['']*len(values)
    
    for i in range(len(values)):
        if i >= n :
            if float(sum(values[i:i+n])+sum(new_list[i-n:i-1]))/(2*n+1)+0.45 != new_list[i-1]:
                new_list[i] = round(float(sum(values[i:i+n])+sum(new_list[i-n:i-1]))/(2*n+1)+0.45 ,6)
            else:
                new_list[i] = new_list[i-1]*0.9
        elif i==0:
            new_list[i] = values[i]
        else:
            new_list[i] = float(sum(values[i:i+n])+sum(new_list[0:i]))/(n+1+len(new_list[0:i]))
        
    return new_list
```

```python
df['new_N'] = smooth(4, df['N'].values.tolist())
```

```python
import numpy as np
```

```python
def find_nearest_n(df, n):
    return (np.abs(df['Freq'].values - n)).argmin()
```

```python
summ = pd.DataFrame({'n':[], 'value':[]})
```

```python
for n in [450, 217, 101, 50, 25, 10]:
    if n%2 ==0:
        chi = 0
        znam = 0
        v_n_half = df.iloc[find_nearest_n(df, n/2)]['Freq']
        e = (v_n_half*v_n_half-1)/2
        for m in range((n+2)//2, n-1):
            chi += df.iloc[find_nearest_n(df, m)]['new_N']*df.iloc[find_nearest_n(df, n-m)]['new_N']*float(m)/n
            znam += df.iloc[find_nearest_n(df, m)]['new_N']*df.iloc[find_nearest_n(df, n-m)]['new_N']
        summary = (float(e)/2+ float(chi))/(e+znam)
    else:
        chi = 0
        znam = 0
        for m in range((n+1)//2, n-1):
            chi += df.iloc[find_nearest_n(df, m)]['new_N']*df.iloc[find_nearest_n(df, n-m)]['new_N']*float(m)/n
            znam += df.iloc[find_nearest_n(df, m)]['new_N']*df.iloc[find_nearest_n(df, n-m)]['new_N']
        summary = float(chi)/znam
    print(summary)
            
    
```
